Importaremos o matplotlib e o numpy

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

Agora importaremos o Glob para ter acesso a todos os arquivos da pasta dados

```python
import glob
```

```python
arquivos = glob.glob('dados/*.txt')
```

```python
print(arquivos)
```

```python
for nome_arquivo in arquivos:
    print(nome_arquivo)
```

Para cada arquivo numa lista de arquivos, carregaremos os dados do arquivo, definiremos as variaveis com base nos dados e plotaremos grafico com essas variaveis

```python
for arquivo in arquivos:
    dados = np.loadtxt(arquivo, comments='%') 
    anos = dados[:, 0]
    meses = dados[:, 1]
    anomalia_temperatura = dados[:, 4]
    ano_decimal = (meses - 1)/12 + anos
    plt.figure()
    plt.plot(ano_decimal, anomalia_temperatura, '-k')
    plt.xlabel('Ano')
    plt.ylabel('Anomalia de Temperatura(°C)')
    
```
